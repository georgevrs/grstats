```python
import pandas as pd

df = pd.read_excel("assets/NFG/A0701_SEL05_TS_QQ_01_1999_01_2025_01E_F_BI.xlsx")
```

```python
# find all sheets in the file and open them in different dataframes
sheet_names = pd.ExcelFile("assets/NFG/A0701_SEL05_TS_QQ_01_1999_01_2025_01E_F_BI.xlsx").sheet_names
dfs = {sheet: pd.read_excel("assets/NFG/A0701_SEL05_TS_QQ_01_1999_01_2025_01E_F_BI.xlsx", sheet_name=sheet) for sheet in sheet_names}

#rename sheets to sheet1 and so on
dfs_renamed = {f"sheet{i+1}": df for i, (sheet, df) in enumerate(dfs.items())}

df1 = dfs_renamed["sheet1"]
df2 = dfs_renamed["sheet2"]

```

```python
# in df1 remove first 2 rows
# rename first cell to "Date"
# make first row the header
df1 = df1.iloc[2:]
df1.columns = df1.iloc[0]
df1 = df1[1:]
df1.rename(columns={df1.columns[0]: "Date"}, inplace=True)
#remove first row
df1 = df1.iloc[1:]
# in column Date remove all *
df1["Date"] = df1["Date"].str.replace("*", "", regex=False)
# in column Date keep only the rows with format YYYY-QX
df1 = df1[df1["Date"].str.match(r"^\d{4}-Q[1-4]$")]
# RESET index
df1.reset_index(drop=True, inplace=True)

df1.columns = df1.columns.str.strip()
# print unique values in column "D995"
#print columns
print(df1.columns)
# in row index 0 insert new row with all values as "_Z"
df1.loc[-1] = ["_Z"] * len(df1.columns)  # adding a row
df1.index = df1.index + 1  # shifting index
df1 = df1.sort_index()  # sorting by index
#cell [0,0] to "INSTR_ASSET"
df1.iloc[0, 0] = "INSTR_ASSET"

#in row index 0 insert new row with all values as header by copy them
columns = df1.columns.tolist()
df1.loc[-1] = columns  # adding a row with column names
df1.index = df1.index + 1  # shifting index
df1 = df1.sort_index()  # sorting by index
df1.iloc[0, 0] = "NA_ITEM"

#same task for df2
df2 = df2.iloc[1:]
# in ALL first row split by "\n" and take the first part
df2.columns = df2.iloc[0].str.split("\n").str[0]
#remove first row
df2 = df2[1:]
#remove row index 2
df2 = df2.drop(df2.index[1])
# edit cell[0,0] to INSTR_ASSET
df2.iloc[0, 0] = "INSTR_ASSET"

df2.loc[-1] = ["_Z"] * len(df2.columns)  # adding a row
df2.index = df2.index + 1  # shifting index
df2 = df2.sort_index()  # sorting by index
#cell [0,0] to "INSTR_ASSET"
df2.iloc[0, 0] = "NA_ITEM"

#strip all column names in df2
df2.columns = df2.columns.str.strip()

#rename columb Quarter to Date

#rename column "Quarter" to "Date"
df2.rename(columns={"Quarter": "Date"}, inplace=True)

df2["Date"] = df2["Date"].str.replace("*", "", regex=False)
# in column Date keep only the rows with format YYYY-QX after row 2
# Regex pattern for YYYY-QX format
pattern = r'^\d{4}-Q[1-4]$'
print(df2.columns)

# Keep first two rows unchanged, then filter the rest
df2 = pd.concat([
    df2.iloc[:2],  # Keep the first two rows
    df2.iloc[2:][df2.iloc[2:]['Date'].str.match(pattern, na=False)]  # Filter from row 2 onward
]).reset_index(drop=True)

# RESET index
df2.reset_index(drop=True, inplace=True)
```

```python
#merge df1 and df2 on column "Date"
merged_df = pd.merge(df1, df2, on="Date", suffixes=('_df1', '_df2'))
# add row index 0 with all values as "_Z"
merged_df.loc[-1] = ["_Z"] * len(merged_df.columns)
merged_df.index = merged_df.index + 1  # shifting index
merged_df = merged_df.sort_index()  # sorting by index
#cell [0,0] to "MATURITY"
merged_df.iloc[0, 0] = "MATURITY"
# at column "AF.31 Short-term debt securities" change first row value to "S"
merged_df.iloc[0, merged_df.columns.get_loc("AF.31 Short-term debt securities")] = "S"
# at column "AF.32 Long-term debt securities" change first row value to "L"
merged_df.iloc[0, merged_df.columns.get_loc("AF.32 Long-term debt securities")] = "L"
# at column "AF.41 Short-term loans" change first row value to "S"
merged_df.iloc[0, merged_df.columns.get_loc("AF.41 Short-term loans")] = "S"
# at column "AF.42 Long-term loans" change first row value to "L"
merged_df.iloc[0, merged_df.columns.get_loc("AF.42 Long-term loans")] = "L"

# add column at index 1 with all values as "Q" name it FREQ
merged_df.insert(1, "FREQ", "Q")
```
